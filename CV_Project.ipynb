```python
import os
import torch
import torch.nn as nn
import torch.utils.data as data
import torchvision.transforms as transforms
import xml.etree.ElementTree as ET
from PIL import Image
from tqdm import tqdm

# --- Model, Loss, and Dataset remain the same ---

class YOLOv1(nn.Module):
    def __init__(self, S=7, B=2, C=20):
        super(YOLOv1, self).__init__()
        self.S = S
        self.B = B
        self.C = C

        self.conv_layers = nn.Sequential(
            nn.Conv2d(3, 64, 7, 2, 3), nn.LeakyReLU(0.1),
            nn.MaxPool2d(2, 2),
            nn.Conv2d(64, 192, 3, 1, 1), nn.LeakyReLU(0.1),
            nn.MaxPool2d(2, 2),
            nn.Conv2d(192, 128, 1), nn.LeakyReLU(0.1),
            nn.Conv2d(128, 256, 3, 1, 1), nn.LeakyReLU(0.1),
            nn.Conv2d(256, 256, 1), nn.LeakyReLU(0.1),
            nn.Conv2d(256, 512, 3, 1, 1), nn.LeakyReLU(0.1),
            nn.MaxPool2d(2, 2),
            *[nn.Sequential(
                nn.Conv2d(512, 256, 1), nn.LeakyReLU(0.1),
                nn.Conv2d(256, 512, 3, 1, 1), nn.LeakyReLU(0.1),
            ) for _ in range(4)],
            nn.Conv2d(512, 512, 1), nn.LeakyReLU(0.1),
            nn.Conv2d(512, 1024, 3, 1, 1), nn.LeakyReLU(0.1),
            nn.MaxPool2d(2, 2),
            nn.Conv2d(1024, 512, 1), nn.LeakyReLU(0.1),
            nn.Conv2d(512, 1024, 3, 1, 1), nn.LeakyReLU(0.1),
            nn.Conv2d(1024, 512, 1), nn.LeakyReLU(0.1),
            nn.Conv2d(512, 1024, 3, 1, 1), nn.LeakyReLU(0.1),
            nn.Conv2d(1024, 1024, 3, 2, 1), nn.LeakyReLU(0.1),
            nn.Conv2d(1024, 1024, 3, 1, 1), nn.LeakyReLU(0.1),
            nn.Conv2d(1024, 1024, 3, 1, 1), nn.LeakyReLU(0.1),
        )

        self.fc_layers = nn.Sequential(
            nn.Flatten(),
            nn.Linear(1024 * 7 * 7, 4096),
            nn.Dropout(0.5),
            nn.LeakyReLU(0.1),
            nn.Linear(4096, S * S * (C + B * 5))
        )

    def forward(self, x):
        x = self.conv_layers(x)
        x = self.fc_layers(x)
        x = x.view(-1, self.S, self.S, self.C + self.B * 5)
        return x

class YoloLoss(nn.Module):
    def __init__(self, S=7, B=2, C=20, lambda_coord=5, lambda_noobj=0.5):
        super(YoloLoss, self).__init__()
        self.mse = nn.MSELoss(reduction="sum")
        self.S = S
        self.B = B
        self.C = C
        self.lambda_coord = lambda_coord
        self.lambda_noobj = lambda_noobj

    def forward(self, predictions, target):
        N = predictions.size(0)
        coord_mask = target[..., 4] > 0
        noobj_mask = target[..., 4] == 0

        coord_pred = predictions[coord_mask]
        coord_target = target[coord_mask]

        coord_loss = self.mse(coord_pred[..., 0], coord_target[..., 0]) + \
                     self.mse(coord_pred[..., 1], coord_target[..., 1]) + \
                     self.mse(torch.sqrt(torch.abs(coord_pred[..., 2]) + 1e-6), torch.sqrt(torch.abs(coord_target[..., 2]) + 1e-6)) + \
                     self.mse(torch.sqrt(torch.abs(coord_pred[..., 3]) + 1e-6), torch.sqrt(torch.abs(coord_target[..., 3]) + 1e-6))

        obj_loss = self.mse(coord_pred[..., 4], coord_target[..., 4])

        noobj_loss = self.mse(predictions[noobj_mask][..., 4], target[noobj_mask][..., 4])

        class_loss = self.mse(coord_pred[..., 5:], coord_target[..., 5:])

        total_loss = self.lambda_coord * coord_loss + obj_loss + \
                     self.lambda_noobj * noobj_loss + class_loss

        return total_loss / N

class VOCDataset(data.Dataset):
    def __init__(self, image_dir, label_dir, S=7, B=2, C=20, transform=None):
        self.image_dir = image_dir
        self.label_dir = label_dir
        self.S = S
        self.B = B
        self.C = C
        self.transform = transform
        self.images = [img for img in os.listdir(image_dir) if img.endswith(".jpg")]
        self.class_map = self.load_class_map()

    def load_class_map(self):
        return {
            "aeroplane": 0, "bicycle": 1, "bird": 2, "boat": 3, "bottle": 4,
            "bus": 5, "car": 6, "cat": 7, "chair": 8, "cow": 9,
            "diningtable": 10, "dog": 11, "horse": 12, "motorbike": 13,
            "person": 14, "pottedplant": 15, "sheep": 16, "sofa": 17,
            "train": 18, "tvmonitor": 19
        }

    def __len__(self):
        return len(self.images)

    def __getitem__(self, idx):
        img_path = os.path.join(self.image_dir, self.images[idx])
        label_path = os.path.join(self.label_dir, self.images[idx].replace(".jpg", ".xml"))
        image = Image.open(img_path).convert("RGB")
        boxes = self.parse_xml(label_path)

        if self.transform:
            image = self.transform(image)

        label_matrix = torch.zeros((self.S, self.S, self.C + self.B * 5))
        for box in boxes:
            x_center, y_center, width, height, class_idx = box
            i, j = int(self.S * y_center), int(self.S * x_center)

            if i >= self.S:
                i = self.S - 1
            if j >= self.S:
                j = self.S - 1

            x_cell, y_cell = self.S * x_center - j, self.S * y_center - i
            w_cell, h_cell = width * self.S, height * self.S

            if label_matrix[i, j, 4] == 0:
                label_matrix[i, j, 0:4] = torch.tensor([x_cell, y_cell, w_cell, h_cell])
                label_matrix[i, j, 4] = 1
                label_matrix[i, j, 5 + class_idx] = 1

        return image, label_matrix

    def parse_xml(self, path):
        tree = ET.parse(path)
        root = tree.getroot()
        boxes = []
        for obj in root.findall("object"):
            class_name = obj.find("name").text
            class_idx = self.class_map[class_name]
            bndbox = obj.find("bndbox")
            xmin = int(bndbox.find("xmin").text)
            ymin = int(bndbox.find("ymin").text)
            xmax = int(bndbox.find("xmax").text)
            ymax = int(bndbox.find("ymax").text)
            x_center = ((xmin + xmax) / 2) / 448
            y_center = ((ymin + ymax) / 2) / 448
            width = (xmax - xmin) / 448
            height = (ymax - ymin) / 448
            boxes.append([x_center, y_center, width, height, class_idx])
        return boxes

# Configuration
image_dir = "/content/VOCdevkit/VOC2007/JPEGImages"
label_dir = "/content/VOCdevkit/VOC2007/Annotations"
epochs = 10
batch_size = 8
lr = 1e-4

transform = transforms.Compose([
    transforms.Resize((448, 448)),
    transforms.ToTensor(),
])

dataset = VOCDataset(image_dir=image_dir, label_dir=label_dir, transform=transform)
loader = torch.utils.data.DataLoader(dataset, batch_size=batch_size, shuffle=True, drop_last=True)

model = YOLOv1()
criterion = YoloLoss()
optimizer = torch.optim.Adam(model.parameters(), lr=lr)

for epoch in range(epochs):
    model.train()
    epoch_loss = 0
    loop = tqdm(loader, desc=f"Epoch {epoch+1}/{epochs}")
    for imgs, labels in loop:
        preds = model(imgs)
        loss = criterion(preds, labels)
        epoch_loss += loss.item()
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        loop.set_postfix(loss=loss.item())

    print(f"Epoch [{epoch+1}/{epochs}] Loss: {epoch_loss / len(loader):.4f}")

torch.save(model.state_dict(), "yolov1.pth")
```
