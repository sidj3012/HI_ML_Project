```python
import pandas as pd
from huggingface_hub import InferenceClient

# Initialize the InferenceClient with your Hugging Face API key
client = InferenceClient(api_key="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")

# Load the CSV containing the English text
input_csv = 'english.csv'  # Replace with the correct path to your CSV
output_csv = 'mistral_translations.csv'

# Read the input CSV into a pandas DataFrame
df = pd.read_csv(input_csv)

# Ensure the column for English text is named 'english_text' (if not, adjust accordingly)
english_texts = df['english_text'].tolist()

# List to store the translations
translated_french = []

# Iterate over each English sentence in the CSV
for text in english_texts:
    # Get the translation from the model
    translated = ""
    for message in client.chat_completion(
        model="mistralai/Mistral-7B-Instruct-v0.3",
        messages=[{"role": "user", "content": f"Translate this to French: {text}"}],
        max_tokens=100,
        stream=True
    ):
        translated += message.choices[0].delta.content
    
    print(translated)
    # Append the translated French text to the list
    translated_french.append(translated.strip())

# Add the translations to the DataFrame
df['translated_french'] = translated_french

# Save the updated DataFrame to a new CSV
df.to_csv(output_csv, index=False)

print(f"Translations saved to {output_csv}")
```
