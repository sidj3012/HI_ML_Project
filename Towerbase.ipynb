```python
import torch
from transformers import pipeline

# Load the translation model pipeline (this assumes Unbabel/TowerInstruct-7B-v0.2 supports translation)
pipe = pipeline("text-generation", model="Unbabel/TowerInstruct-7B-v0.2", torch_dtype=torch.bfloat16, device_map="auto")

# Sample dataset (ds['validation'] is assumed to be loaded elsewhere)
# Assuming ds['validation'] is a Hugging Face dataset or similar structure
# Replace this with actual loading mechanism of the dataset
# For example: ds = load_dataset("your_dataset_name")
# and then: validation_data = ds['validation']

# Simulating the dataset structure
# In reality, this will come from your dataset (ds['validation'])
validation_data = {"text": ["The researchers launched a new model.", "Artificial intelligence is transforming industries."]}

# List to store the French translations
french_translations = []

# Iterate over each text in the validation dataset
for text in validation_data['text']:
    # Prepare the prompt for translation
    messages = [
        {"role": "user", "content": f"Translate the following text from English to French.\nEnglish: {text}\nFrench:"},
    ]

    # Apply the chat template and generate the prompt
    prompt = pipe.tokenizer.apply_chat_template(messages, tokenize=False, add_generation_prompt=True)

    # Use the model to translate the text
    outputs = pipe(prompt, max_new_tokens=256, do_sample=False)

    # Extract and print the generated French text
    translation = outputs[0]["generated_text"]
    print(f"English: {text}\nFrench: {translation}\n")

    # Append to the list of translations
    french_translations.append(translation)

# Optionally, you can add the translations back to your dataset or save them for later use.
```

```python
# Install necessary libraries
!pip install datasets huggingface_hub transformers accelerate

# Import the required libraries
from transformers import AutoModelForSeq2SeqLM, AutoTokenizer, pipeline
from datasets import load_dataset
import torch

# Load the dataset (replace 'your_dataset_name' with the actual name)
ds = load_dataset("hf-internal-testing/librispeech_asr_dummy")

# Function to translate English text to French using the transformers pipeline
def translate_to_french(text):
    """Translates English text to French using the 'translation_en_to_fr' pipeline."""
    # Load the translation model and tokenizer (if not already loaded)
    model_name = "Helsinki-NLP/opus-mt-en-fr"  # Use a smaller, publicly available translation model
    model = AutoModelForSeq2SeqLM.from_pretrained(model_name)
    tokenizer = AutoTokenizer.from_pretrained(model_name)

    # Create the translation pipeline
    translator = pipeline("translation_en_to_fr", model=model, tokenizer=tokenizer, device_map="auto")

    # Translate the text
    translation = translator(text, max_length=500)[0]['translation_text']

    return translation

# Loop over the 'validation' set and translate each text row
translated_texts = []
for i, text in enumerate(ds['validation']['text']):  # Assuming the column is named 'text'
    print(f"Translating row {i+1}: {text}")
    translation = translate_to_french(text)
    translated_texts.append(translation)
    print(f"French translation: {translation}\n")

# (Optional) Display the first 5 translations
for i, translation in enumerate(translated_texts[:5]):
    print(f"English: {ds['validation']['text'][i]}")
    print(f"French: {translation}\n")
```

```python
import pandas as pd

# Load TowerBase_french.csv
TowerBase_df = pd.read_csv("/content/TowerBase_French.csv")

# Load fr_text.xlsx
gold_standard_df = pd.read_excel("/content/fr_text.xlsx")

# Assuming that the CSV and Excel files have the same column names
# Column for TowerBase translations - CHANGED TO 'TowerBase_French'
TowerBase_translations = TowerBase_df['TowerBase_French'].tolist()

# Column for gold standard translations
gold_standard_translations = gold_standard_df['texte'].tolist()
```

##BLEU Score

```python
import sacrebleu

# Calculate BLEU score
bleu_score = sacrebleu.corpus_bleu(TowerBase_translations, [gold_standard_translations])

print(f"BLEU score: {bleu_score.score}")
```

##CoMET Score

```python
from comet import download_model, load_from_checkpoint

# Download the COMET model
model_path = download_model("Unbabel/wmt20-comet-da")  # Or use another model if you prefer
model = load_from_checkpoint(model_path)

# Prepare data for COMET
data = [
    {"src": "", "mt": TowerBase_translations[i], "ref": gold_standard_translations[i]}
    for i in range(len(TowerBase_translations))
]

# Compute COMET scores
comet_scores = model.predict(data, batch_size=8, gpus=1)

# Calculate the average COMET score
average_comet_score = sum(comet_scores["scores"]) / len(comet_scores["scores"])

print(f"Average COMET score: {average_comet_score}")
```
